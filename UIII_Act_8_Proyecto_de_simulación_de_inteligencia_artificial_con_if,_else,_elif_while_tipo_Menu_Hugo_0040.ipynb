#### Albañil y ladrillo



```python
# Simulación de IA para Asistente de Albañilería
nombre_maestro = input("Ingrese el nombre del maestro albañil: ")
ladrillos_en_almacen = 1000
obra_activa = True

while obra_activa:
    print("\n--- MENÚ DE OBRA (Maestro:", nombre_maestro, ") ---")
    print("1. Calcular ladrillos | 2. Retirar material | 3. Evaluación de riesgo | 4. Salir")
    opcion = input("Seleccione una opción (1-4): ")

    if opcion == "1":
        # Uso de operadores aritméticos para calcular área y materiales [3]
        largo = float(input("Largo del muro (metros): "))
        alto = float(input("Alto del muro (metros): "))
        # Se estiman 60 ladrillos por metro cuadrado
        total_necesario = largo * alto * 60
        print(f"IA: Para un muro de {largo*alto}m2, necesita aproximadamente {total_necesario} ladrillos.")

    elif opcion == "2":
        # Validación con operadores relacionales [4, 5]
        cantidad = int(input("¿Cuántos ladrillos desea retirar del almacén?: "))
        if cantidad <= ladrillos_en_almacen and cantidad > 0:
            ladrillos_en_almacen -= cantidad # Resta acumulativa [3]
            print(f"IA: Retiro exitoso. Quedan {ladrillos_en_almacen} ladrillos en stock.")
        else:
            print("IA: Error. Cantidad inválida o stock insuficiente.")

    elif opcion == "3":
        # Lógica de decisión jerárquica con if-elif [5, 6]
        altura_muro = float(input("¿A qué altura (metros) está construyendo?: "))
        if altura_muro > 3.0:
            print("IA: ¡ALERTA! Altura peligrosa. Se requiere andamio reforzado y arnés.")
        elif altura_muro >= 1.5:
            print("IA: Precaución. Altura considerable, use casco y andamio estándar.")
        else:
            print("IA: Altura segura para trabajar desde el suelo.")

    elif opcion == "4":
        # Condición de salida para romper el ciclo [7, 8]
        print("IA: Finalizando asistencia. ¡Que tenga una buena jornada,", nombre_maestro, "!")
        obra_activa = False

    else:
        print("IA: Opción no reconocida. Intente de nuevo.")
```
